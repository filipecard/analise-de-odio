### 2- Limpeza do texto

```python
import re
import string
```

>Criação das regexes

```python
def clean_text_round1(text):
    text = text.lower()
    text = re.sub(r'@(\w+)\W', ' ', text) #Substituir menções de usuários
    text = re.sub(r'https?://\S*', ' ', text) #Remover links
    text = re.sub(r'&[^;.]*;', ' ', text) #Apagar elementos de html    
    text = re.sub('[%s]' % string.punctuation, " ", text)
    #text = re.sub('[^\w\s\d]', ' ', text)
    text = re.sub('\s{2,}', ' ', text) #Remover excesso de espaços
    text = re.sub(r'\s', ' ', text) #Remover whitespaces    
    return text

round1 = lambda x: clean_text_round1(x)
```

>Aplicação da limpeza

```python
clean = data.drop(columns='tweet_text')
clean['tweet_text'] = pd.DataFrame.from_dict(data.tweet_text.apply(round1))
clean['length'] = [len(text) for text in clean.tweet_text]
```

### Tokenização

```python
corpus = " ".join([text for text in clean.tweet_text])
```

```python
import nltk
```

```python
words = nltk.word_tokenize(corpus)
```

```python
words
```

### Stemming

```python
from nltk.stem import PorterStemmer

words_stem = [PorterStemmer().stem(word) for word in words]
```
