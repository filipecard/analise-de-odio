### 2- Limpeza do texto

```python
import re
import string
```

> Coverter encoding

```python
covert_to_ascii = lambda x : x.encode(encoding="ascii",errors="replace").decode()
```

```python
data.tweet_text = pd.DataFrame.from_dict(data.tweet_text.apply(covert_to_ascii))
```

```python
data
```

>Criação das regexes

```python
def clean_text_round1(text):
    punctuation = string.punctuation
    punctuation = re.sub("#@'", "", punctuation)
    text = text.lower()
    text = re.sub(r'@(\w+)\W', 'XmentionX ', text) #Substituir menções de usuários
    text = re.sub(r'https?://\S*', 'XlinkX ', text) #Remover links
    text = re.sub(r"(\w)'(\w)", r"\1\2", text)
    text = re.sub(r'&[^;.]*;', ' ', text) #Apagar elementos de html    
    text = re.sub('[%s]' % punctuation, " ", text) # Remove pontuações
    text = re.sub(r'(.)\1{3,}', r'\1\1', text) # Ajusta repetições
    text = re.sub('\s{2,}', ' ', text) #Remover excesso de espaços
    text = re.sub(r'\s', ' ', text) #Remover whitespaces    
    return text

round1 = lambda x: clean_text_round1(x)
```

>Aplicação da limpeza

```python
clean = data.drop(columns='tweet_text')
clean['tweet_text'] = pd.DataFrame.from_dict(data.tweet_text.apply(round1))
clean['length'] = [len(text) for text in clean.tweet_text]
```
